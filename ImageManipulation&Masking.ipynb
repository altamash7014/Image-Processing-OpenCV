```python
import cv2 as cv

cam = cv.VideoCapture(0)
cv.namedWindow("Frame", cv.WINDOW_NORMAL) 
while True:
    ret, img = cam.read()
    if not ret:
        break

    img = cv.flip(img, 1)
    img = cv.rectangle(img,(160,60),(480,420),(10,200,300),5)
    img = cv.circle(img, (300,250), 200, (108,20,174), 3)
    img = cv.putText(img,"Altamash",(140,260),cv.FONT_HERSHEY_PLAIN,1,(160,67,92),3)
    img = cv.putText(img,"Altamash",(180,240),
                 cv.FONT_HERSHEY_PLAIN,2,(0,0,0),6)
    img = cv.putText(img,"Altamash",(180,240),
                 cv.FONT_HERSHEY_PLAIN,2,(255,255,255),2)

    cv.imshow("Frame", img)

    if cv.waitKey(10) == 27:
        break

cam.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv

cam = cv.VideoCapture(0)
cv.namedWindow("Frame", cv.WINDOW_NORMAL)

while True:
    ret, img = cam.read()
    if not ret:
        break

    blur1 = cv.blur(img, (4,4))
    blur2 = cv.blur(img, (8,8))
    blur3 = cv.blur(img, (12,12))
    blur4 = cv.blur(img, (15,15))
    final = cv.Canny(blur2,50,100)
    

    cv.imshow("blur1", blur1)
    cv.imshow("blur2", blur2)
    cv.imshow("blur3", blur3)
    cv.imshow("blur4", blur4)
    cv.imshow("Final", final)
    if cv.waitKey(10) == 27:   # ESC
        break

cam.release()
cv.destroyAllWindows()
```

```python
import cv2


img = cv2.imread("image.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Grayscale Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv2.imshow("Binary Threshold", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Binary Inverse", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO)
cv2.imshow("ToZero Threshold", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

camera = cv.VideoCapture(0)
lower = np.array([0, 0, 0])
upper = np.array([100, 100, 100])

while True:
    _, img = camera.read()

    blur = cv.blur(img, (4, 4))
    mask = cv.inRange(blur, lower, upper)

    cv.imshow("Frame", img)
    cv.imshow("Mask", mask)

    if cv.waitKey(1) == 27:
        camera.release()
        break
    
cv.destroyAllWindows()
```
